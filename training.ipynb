```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.iloc[:,:-1],df.iloc[:,-1],test_size=0.1,random_state=42)
```

```python
X_train.shape,y_train.shape
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled=scaler.fit_transform(X_test)
```

```python
import pickle
```

```python
with open("scaler.pkl","wb") as file:
    pickle.dump(scaler,file)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,BaggingClassifier
from xgboost import XGBClassifier
from sklearn.neighbors import KNeighborsClassifier

```

```python
def train_model(x_train,x_test,y_train,y_test):
    scores={}
    models={
        "logistic":LogisticRegression(),
        "knn":KNeighborsClassifier(),
        "dec_tree":DecisionTreeClassifier(),
    }
    
    for key,value in models.items():
        model=value
        model.fit(x_train,y_train)
        #predictions=model.predict(x_test)
        
        scores[key]=model.score(x_test,y_test)
        
        
        
    return scores
```

```python
train_model(X_train_scaled,X_test_scaled,y_train,y_test)
```

```python
def train_slower_models(x_train,x_test,y_train,y_test):
    scores={}
    models={  
        "svc":SVC(),
        "Random_forest":RandomForestClassifier(),
        "Adaboost":AdaBoostClassifier(),
        "Bagging":BaggingClassifier(),
        
    }
    
    for key,value in models.items():
        model=value
        model.fit(x_train,y_train)
        #predictions=model.predict(x_test)
        
        scores[key]=model.score(x_test,y_test)
        
        
        
    return scores
```
